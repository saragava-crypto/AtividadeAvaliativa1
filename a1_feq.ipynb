```python
import numpy as np
#criação do vetor c para armazenar os valores de C fornecidos pelo usuário.
c = []

#uso da estrutura de repetição for considerando 10 leituras consecutivas.
for i in range(10):
  C = float(input("Digite a concentração do reagente, considerando valores entre 0 e 10 mol/L\n")) # dados de C fornecido pelo usuário.
  c.append(C) #prenchimento do vetor c

#contadores.
b = adeq = a = 0

print("\n") #serve para quebra de linha para melhor organização.

#uso da estrutura de repetição for considerando 10 leituras consecutivas mais o uso de estrutura condicional if.
#o uso da estrutura condicional if foi usada para a classificação da concentração além de contabilizar quantas vezes cada condição ocorreu.
for C in c:
    if C < 3:
      b+=1
    elif C <=7:
      adeq+=1
    else:
      a+=1

#impressão do número de ocorrências de cada classe (Baixa, Adequada, Alta).
print("\n")
print("Resumo:")
print("Ocorrências Nível Baixo:",b)
print("Ocorrências Nível Adequado:",adeq)
print("Ocorrências Nível Alto:",a)
print("\n")

#o uso da estrutura condicional if/elif/else para analisar qual condição foi predominante no processo.
if b > adeq and b > a:
  print("Condição predominante: Nivel Baixo!")
elif adeq > b and adeq > a:
  print("Condição predominante: Nivel Adequado!")
else:
  print("Condição predominante: Nível Alto!")

#só não entendi como seria feito o uso da função dentro desse código


```
